```python
import scipy.stats as stats
from scipy.stats import chi2
```

### Performing the chi-squared test

```python
for i in data.keys():
  data_table = pd.crosstab(data['class'], data[i])
  Observed_values = data_table.values
  val = stats.chi2_contingency(data_table)
  Expected_val = val[3]
  no_of_rows = len(Observed_values)
  no_of_columns = len(Observed_values[0])
  dof = (no_of_rows-1)*(no_of_columns-1)
  alpha = 0.05
  chi_square = sum(((o-e)**2)/e for o,e in zip(Observed_values, Expected_val))
  chi_square_stats = sum(chi_square)
  chi_square_stats
  critical_value = chi2.ppf(q=1-alpha, df=dof)
  p_value = 1- chi2.cdf(x = chi_square_stats, df=dof)
  print(i, p_value, alpha, "............", chi_square_stats, critical_value, dof, val)
```

### Dropping 9 unwanted features

###Spliting the data into X and y. (dependent and independent)

```python

X = data1.drop('class', axis=1)
y = data1['class']
X.head()
```

```python
y.head()
```

### Preparing the Data for fitting by one hot encoding and test, train split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train = pd.get_dummies(X_train)
X_train.head()
```

```python
y_train = pd.get_dummies(y_train)
y_train.head()
y_train = y_train.drop('p', axis=1)
y_train.head()
```

###Using Decision Tree

```python
clf = tree.DecisionTreeClassifier(random_state=42)
clf = clf.fit(X_train, y_train)

```

```python
y_test = pd.get_dummies(y_test)
y_test = y_test.drop('p', axis=1)
```

```python
X_test = pd.get_dummies(X_test)
```

```python
pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy_score(y_test, pred)
```
